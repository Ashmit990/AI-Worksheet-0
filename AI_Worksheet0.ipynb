```python
time_data=[
(3.5,2.0,7.0),(5.0,1.5,6.5),(2.5,3.0,8.0),(4.0,2.0,6.0),(1.5,4.5,9.0),
(3.0,2.5,7.5),(5.5,1.0,6.0),(2.0,3.5,8.5),(4.5,2.0,7.0),(3.0,3.0,7.5),
(6.0,1.5,6.0),(2.5,4.0,8.0),(4.0,2.5,7.0),(5.0,2.0,6.5),(3.5,2.5,7.0)
]

low=[]
moderate=[]
high=[]

for s,e,sl in time_data:
    if s<3:
        low.append(s)
    elif 3<=s<=5:
        moderate.append(s)
    else:
        high.append(s)

print(low)
print(moderate)
print(high)
```

```python
print(len(low))
print(len(moderate))
print(len(high))
```

```python
study_minutes=[]
for s,e,sl in time_data:
    study_minutes.append(s*60)
print(study_minutes)
```

```python
study_hours=[]
entertainment_hours=[]
sleep_hours=[]

for s,e,sl in time_data:
    study_hours.append(s)
    entertainment_hours.append(e)
    sleep_hours.append(sl)

print(sum(study_hours)/len(study_hours))
print(sum(entertainment_hours)/len(entertainment_hours))
print(sum(sleep_hours)/len(sleep_hours))
```

```python
import matplotlib.pyplot as plt

study=[s for s,e,sl in time_data]
sleep=[sl for s,e,sl in time_data]

plt.scatter(study,sleep)
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study vs Sleep Pattern")
plt.show()
```

```python
def sum_nested_list(a):
    t=0
    for x in a:
        if isinstance(x,list):
            t+=sum_nested_list(x)
        else:
            t+=x
    return t

print(sum_nested_list([1,[2,[3,4],5],6,[7,8]]))
```

```python
def generate_permutations(s):
    if len(s)==1:
        return [s]
    res=[]
    for i in range(len(s)):
        for p in generate_permutations(s[:i]+s[i+1:]):
            res.append(s[i]+p)
    return list(set(res))

print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def calculate_directory_size(d):
    total=0
    for k,v in d.items():
        if isinstance(v,dict):
            total+=calculate_directory_size(v)
        else:
            total+=v
    return total

directory_structure={
"file1.txt":200,
"file2.txt":300,
"subdir1":{"file3.txt":400,"file4.txt":100},
"subdir2":{"subsubdir1":{"file5.txt":250},"file6.txt":150}
}

print(calculate_directory_size(directory_structure))
```

```python
def min_coins(coins,amount):
    dp=[float("inf")]*(amount+1)
    dp[0]=0
    for c in coins:
        for i in range(c,amount+1):
            dp[i]=min(dp[i],dp[i-c]+1)
    return dp[amount] if dp[amount]!=float("inf") else -1

print(min_coins([1,2,5],11))
```

```python
def longest_common_subsequence(a,b):
    n=len(a)
    m=len(b)
    dp=[[0]*(m+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for j in range(1,m+1):
            if a[i-1]==b[j-1]:
                dp[i][j]=dp[i-1][j-1]+1
            else:
                dp[i][j]=max(dp[i-1][j],dp[i][j-1])
    return dp[n][m]

print(longest_common_subsequence("abcde","ace"))
```

```python
def knapsack(w,v,c):
    n=len(w)
    dp=[[0]*(c+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for cap in range(1,c+1):
            if w[i-1]<=cap:
                dp[i][cap]=max(dp[i-1][cap],v[i-1]+dp[i-1][cap-w[i-1]])
            else:
                dp[i][cap]=dp[i-1][cap]
    return dp[n][c]

print(knapsack([1,3,4,5],[1,4,5,7],7))
```
